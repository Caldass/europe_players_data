```python
import pandas as pd
```

```python
df = pd.read_csv('scraped_df.csv')
```

```python
df.shape
```

```python
#making all columns lowercase and replace white spaces with _
df.columns = df.columns.str.replace(' ', '_').str.lower()
```

```python
#Every stat with the "last" prefix means that its a stat from the previous season
df.columns
```

```python
#using columns that will be usefull to our eda and our model
df = df[['short_name', 'latest_market_value', 'highest_market_value','current_club',
         'position', 'main_position', 'secondary_position',  'age', 'height','citizenship', 'league_country'
         ,'foot', 'player_agent', 'outfitter',  'league', 'on_loan_from', 'contract_expires',
         'appearances', 'goals', 'assists','yellow', 'second_yellow', 'red', 'minutes_played','goals_conceded',
         'clean_sheets','last_appearances', 'last_goals', 'last_yellow','last_second_yellow','last_red', 
         'last_goals_conceded', 'last_clean_sheets', 'last_assists','last_minutes_played']]
```

```python
df.head()
```

```python
to_numeric_cols = ['appearances', 'goals', 'yellow', 'second_yellow', 
    'red', 'goals_conceded', 'clean_sheets', 'last_appearances', 'last_goals', 
    'last_yellow', 'last_second_yellow', 'last_red', 'last_goals_conceded', 'last_clean_sheets', 
    'assists', 'last_assists', 'minutes_played', 'last_minutes_played', 'contract_expires']
```

```python
df[to_numeric_cols] = df[to_numeric_cols].replace('-',0)

#since we will be using this list to convert columns to int type columns, we need to remove the date column
to_numeric_cols.remove('contract_expires')
```

```python
df.shape
```

```python
df.info()
```

```python
df['citizenship'].unique()
```

```python
#Get the citizenship
s = df['citizenship'].str.split(u'\xa0', n=3, expand=True)
df['country'], df['country_2'] = s[0], s[2]
df.drop(columns = ['citizenship'], inplace = True, axis = 1)
```

```python
df.country.unique()
```

```python
df.country_2.unique()
```

```python
#Converting string type currency to a float
df['latest_value_letter'] = df['latest_market_value'].str.extract('([A-Za-z]+)')
df['latest_market_value'] = df['latest_market_value'].str.extract(r'(\d+.\d+)').astype('float')
df['latest_market_value'] = df.apply(lambda x: x.latest_market_value * 1000 if x.latest_value_letter == 'Th' else x.latest_market_value * 1000000, axis = 1)
df.drop(columns = 'latest_value_letter', inplace = True, axis = 1)

df['highest_value_letter'] = df['highest_market_value'].str.extract('([A-Za-z]+)')
df['highest_market_value'] = df['highest_market_value'].str.extract(r'(\d+.\d+)').astype('float')
df['highest_market_value'] = df.apply(lambda x: x.highest_market_value * 1000 if x.highest_value_letter == 'Th' else x.highest_market_value * 1000000, axis = 1)
df.drop(columns = 'highest_value_letter', inplace = True, axis = 1)
```

```python
df[['latest_market_value', 'highest_market_value']]
```

```python
# % of null values in columns
df.isnull().mean().sort_values()
```

```python
#Filtering out rows from columns that can't have nan values
df = df[df['latest_market_value'].notna()]
df = df[df['league'].notna()]
```

```python
#Filling columns with mode
df['foot'] = df['foot'].fillna(df['foot'].mode()[0])
```

```python
#Height to float and filling NAs with the height mean
df['height'] = df['height'].str.extract(r'(\d+.\d+)')
df['height'] = df['height'].str.replace(',','.').astype('float')
df['height'] = df['height'].fillna(df['height'].mean())
```

```python
#If there's information 1 else 0
df[['on_loan_from','secondary_position', 'country_2']] = df[['on_loan_from','secondary_position','country_2']].where(df[['on_loan_from', 'secondary_position','country_2']].isnull(), 1).fillna(0).astype(int)
```

```python
#Filling with unknown where you can't fill with any other value such as mean and mode
df[['outfitter','player_agent']] = df[['outfitter','player_agent']].fillna('unknown')
```

```python
df.isnull().mean().sort_values()
```

```python
#Removing time string from minutes played columns
df['minutes_played'] = df['minutes_played'].str.replace('.','').str.replace("'","")
df['last_minutes_played'] = df['last_minutes_played'].str.replace('.','').str.replace("'","")
```

```python
#Filling NAs in the numeric cols with 0
df[to_numeric_cols] = df[to_numeric_cols].fillna(0)
```

```python
#convert all numeric column to an integer type column
for i in to_numeric_cols:
    try:
        df[i] = df[i].astype(int)
    except ValueError:
        df[i] = df[i].astype('float').astype(int)
```

```python
#Converting columns to date types
df['contract_expires'] = pd.to_datetime(df['contract_expires'], errors = 'ignore')
```

```python
df.isnull().mean().sort_values()
```

Keepers will never have a main position, since their only position is GoalKeeper. But, there are a few players that don't have a main position because of a mistake from the website or because the website could not get the player's main position. As you can see in the cell bellow, it's an irrelevant % of the observations, so we're just going to remove it.

```python
print('''% of observations that are not keepers and have null "main position" : {}'''.format((len(df[(df['position'] != 'Goalkeeper') & (df['main_position'].isnull())]) / len(df))))
```

```python
#Removing the rows that match this condition
df = df.drop(df[(df['position'] != 'Goalkeeper') & (df['main_position'].isnull())].index)
```

```python
#Filling the null values in the main_position column with goalkeeper since all of those rows are keepers data
df['main_position'] = df.apply(lambda x: 'Goalkeeper' if x.position == 'Goalkeeper' else x.main_position, axis = 1)
```

Now I'm going to rename leagues that have the same name, but are from different countries, since <a href = "https://www.transfermarkt.us/"> transfer market</a> gives the same name to the German Bundesliga and the austrian Bundesliga for example.

```python
#creating dataframe with duplicated leagues
duplicated_leagues = df.groupby(['league', 'league_country']).count().reset_index()[['league','league_country']]
duplicated_leagues['duplicated'] = duplicated_leagues.groupby(['league']).transform('count')
duplicated_leagues = duplicated_leagues[duplicated_leagues['duplicated'] > 1]['league'].unique()
duplicated_leagues
```

```python
#If is duplicated then concatenate with country name
df['league'] = df.apply(lambda x: x.league if x.league not in duplicated_leagues else x.league + "_" + x.league_country, axis = 1)
```

```python
df.to_csv('cleaned_df.csv')
```
